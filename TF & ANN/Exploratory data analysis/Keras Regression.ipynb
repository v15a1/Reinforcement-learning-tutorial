```python
df
df = df.drop("id", axis = 1) #Dropping the ID
df["date"] = pd.to_datetime(df["date"])
df["date"]
```

```python
df["year"] = df["date"].apply(lambda date: date.year)
df["month"] = df["date"].apply(lambda date: date.month)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="month", y="price", data = df)
```

```python
df = df.drop("zipcode", axis = 1) #Dropping the Zipcode
```

```python
df = df.drop('date',axis=1)
```

## Create model

```python
X = df.drop("price", axis=1).values
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state= 101)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

###### Getting the number of features

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(19, activation= "relu"))
model.add(Dense(19, activation= "relu"))
model.add(Dense(19, activation= "relu"))
model.add(Dense(19, activation= "relu"))

model.add(Dense(1))
model.compile(optimizer="adam", loss = "mse")
```

```python
# Batch sizes are in powers of 2 
model.fit(
    x = X_train, 
    y = y_train,
    validation_data = (X_test, y_test),
    batch_size = 128,
    epochs = 400
)
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.plot()
# Val_loss curving upwards means you are overfitting
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df["price"].describe()
```

```python
explained_variance_score(y_test, predictions)
```

```python
 plt.scatter(y_test, predictions)
```

```python
new_house = df.drop("price", axis=1).iloc[0]
new_house = scaler.transform(new_house.values.reshape((-1, 19)))


predicted_val = model.predict(new_house)
actual_val = df["price"][0]

print(predicted_val)
print(actual_val)
```
